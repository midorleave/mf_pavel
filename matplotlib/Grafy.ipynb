```python
import matplotlib.pyplot as plt

option_votes = ([66, 22, 12])
option_names = [
    'Something big',
    'Something smaller',
    'Something small'
]

figure = plt.figure()
axes = figure.add_subplot()

axes.pie(
    option_votes,
    labels = option_names,
    explode = [0.1, 0.1, 0.1],
    autopct = '%1.1f%%'
)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

companies = ['Apple', 'Microsoft', 'FB', 'Amazon']
q1_2020 = [11.2, 10.8, 4.9, 2.5]
q1_2021 = [23.6, 15.5, 9.5, 8.1]

x = np.arange(len(companies))
width = 0.4

fig, ax = plt.subplots(figsize=(8, 8))

q2021 = ax.bar(x - width/2, q1_2021, width, label = 'Q1-2021')
q2020 = ax.bar(x + width/2, q1_2020, width, label = 'Q1-2020')
# q2021 = ax.bar(x, q1_2021, width, label = 'Q1-2021')

plt.title('Q1 2020 vs Q1 2021')
plt.ylabel('Revenue')
ax.set_xticks(x)
ax.set_xticklabels(companies)
ax.legend()

ax.bar_label(q2021, padding = 3)
ax.bar_label(q2020, padding = 3)

plt.show()
```
